```python
"""
Exponencial Growth Model / Modelo de Crescimento Exponencial

Author / Autor: imoripaolo

Based of / Baseado em: Boyce & DiPrima 2.5

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Exponencial Model: dynamics of a population

y' = ry

r > 0: population growth rate
---
Modelo de Exponencial: dinâmica de uma população

y' = ry

r > 0: taxa de crescimento da população

"""

r, T  = 0.4, 100
y = np.linspace(0,100,T)

def f(y):
    f = r*y
    return f

dot_y = np.zeros(T)
for i in range(0,T):
    dot_y[i] = f(y[i])

fig, ax = plt.subplots()
ax.plot(y,dot_y,'b')
plt.axhline(0, color='k', linestyle='-')
ax.set_ylabel('f(y)')
ax.set_xlabel('y')
plt.show()  

# print("As we can see, population always grows")
print("Como podemos ver, a população sempre cresce")
```
