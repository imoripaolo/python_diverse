```python
"""
Growth Model with Threshold / Modelo de Crescimento com Limiar

Author / Autor: imoripaolo

Based of / Baseado em: Boyce & DiPrima 2.5

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Threshold Model: dynamics of a population. If populations falls beyond certain threshold, it goes extinct.

y' = r(1-y/k)y

r > 0: population growth rate

T > 0: threshold
---
Modelo com Limiar: dinâmica de uma população. Se a população cai além de certo limiar, ela será extinta.

y' = r(1-y/k)y

T > 0: Limiar

"""

r, T, n = 0.4, 80, 101
y = np.linspace(0,100,n)

def f(y):
    f = -r*(1-(y/T))*y
    return f

dot_y = np.zeros(n)
for i in range(0,n):
    dot_y[i] = f(y[i])

fig, ax = plt.subplots()
ax.plot(y,dot_y,'b')
plt.axhline(0, color='k', linestyle='-')
ax.set_ylabel('f(y)')
ax.set_xlabel('y')
plt.show() 

# print("Population goes extinct if y < ", y[np.argmax(dot_y)], "= T")
print("População é extinta caso y < ", y[np.argmax(dot_y)], "= T")
```

```python
"""
Initial Value Problem

We'll check how y evolves from different y_0

---
Problema do Valor Inicial

Vamos verificar como y a partir de diferentes y_0

"""
t, m  = 30, 15
y0 = np.linspace(0,1.00005*T,m)
y = np.zeros((t,m))

def g(y):
    g = (1-r*(1-(y/T)))*y
    return g

y[0] = y0

for i in range(1,t):
    for j in range(0,m):
        y[i][j] = g(y[i-1][j])

fig, ax = plt.subplots()
ax.plot(y,'b')
plt.axhline(T, color='k', linestyle='-')
ax.set_ylabel('y')
ax.set_xlabel('t')
#plt.ylim([0, 1.4])
plt.show()
```
