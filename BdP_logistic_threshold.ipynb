```python
"""
Logistic Growth Model with Threshold / Modelo de Crescimento Logístico com Limiar

Author / Autor: imoripaolo

Based of / Baseado em: Boyce & DiPrima 2.5

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Logistic Model with Threshold: dynamics of a population. If populations falls beyond certain threshold, it goes extinct.
Otherwise, it stabilizes at certain point

y' = -r(1-y/T)(1-y/k)y

r > 0: population growth rate

K > 0: carrying capacity

T > 0: threshold
---
Modelo Logístico com Limiar: dinâmica de uma população. Se a população cai além de certo limiar, ela será extinta.
Caso contrário, ela estabiliza em algum ponto

y' = -r(1-y/T)(1-y/k)y

K > 0: carrying capacity

T > 0: Limiar

"""

r, T, K, n = 0.4, 60, 100, 101
y = np.linspace(0,100,n)

def f(y):
    f = -r*(1-(y/T))*(1-(y/K))*y
    return f

dot_y = np.zeros(n)
for i in range(0,n):
    dot_y[i] = f(y[i])

fig, ax = plt.subplots()
ax.plot(y,dot_y,'b')
plt.axhline(0, color='k', linestyle='-')
ax.set_ylabel('f(y)')
ax.set_xlabel('y')
plt.show() 

```
