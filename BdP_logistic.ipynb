```python
"""
Logistic Growth Model / Modelo de Crescimento Logístico

Author / Autor: imoripaolo

Based of / Baseado em: Boyce & DiPrima 2.5

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Logistic Model: dynamics of a population. There's a limit on how much populations can grow.

y' = r(1-y/k)y

r > 0: population growth rate

K > 0: carrying capacity
---
Modelo Logístico: dinâmica de uma população. Existe um limite no quanto a população pode crescer.

y' = r(1-y/k)y

K > 0: carrying capacity
"""

r, K, n = 0.4, 80, 101
y = np.linspace(0,100,n)

def f(y):
    f = r*(1-(y/K))*y
    return f

dot_y = np.zeros(n)
for i in range(0,n):
    dot_y[i] = f(y[i])

fig, ax = plt.subplots()
ax.plot(y,dot_y,'b')
plt.axhline(0, color='k', linestyle='-')
ax.set_ylabel('f(y)')
ax.set_xlabel('y')
plt.show() 

# print("Max growth at y = ", y[np.argmax(dot_y)], "= k/2")
print("Maior crescimento em y = ", y[np.argmax(dot_y)], "= k/2")
```

```python
"""
Initial Value Problem

We'll check how y evolves from different y_0

---
Problema do Valor Inicial

Vamos verificar como y a partir de diferentes y_0

"""
T, m  = 30, 15
y0 = np.linspace(0,3*K/2,m)
y = np.zeros((T,m))

def g(y):
    g = r*(1-(y/K))*y + y
    return g

y[0] = y0

for i in range(1,T):
    for j in range(0,m):
        y[i][j] = g(y[i-1][j])

fig, ax = plt.subplots()
ax.plot(y,'b')
plt.axhline(K, color='k', linestyle='-')
ax.set_ylabel('Y(E)')
ax.set_xlabel('E')
plt.show()
```
